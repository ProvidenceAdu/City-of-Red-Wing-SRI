### Hexagons

```python
municapalboundary = os.path.join(projectGDB,'MuniBoundPlusSectRW')
redwingh3hexagon = os.path.join(projectGDB,'RedWingH3')
redwingprojecttemp = os.path.join(projectGDB, 'RedWingH3_prj'),
print("Input FC exists:", arcpy.Exists(municapalboundary))
```

```python
h3hexagons = set()
spatialreference=arcpy.SpatialReference(4326)


with arcpy.da.SearchCursor(municapalboundary, ["SHAPE@"]) as cursor:
    for row in cursor:
        geometry = row[0]
        
        # Project to WGS84 (H3 requires lat/lon)
        geomproject = geometry.projectAs(arcpy.SpatialReference(4326))
        
        # Handle multipart polygons
        for part in geomproject:

            coords = []
            
            for point in part:
                if point:
                    
                    coords.append((point.Y, point.X)) 
            
            if len(coords) > 2:
                poly = h3.LatLngPoly(coords)
                cells = h3.polygon_to_cells(poly, 8)
                h3hexagons.update(cells)

print(f"Generated {len(h3hexagons)} hexagons")
```

```python
arcpy.management.CreateFeatureclass(projectGDB,'RedWingH3HexagonRes7',"POLYGON",None,None,None,spatialreference)

arcpy.management.AddField('RedWingH3HexagonRes7', "H3_ID", "TEXT", field_length=255)

with arcpy.da.InsertCursor('RedWingH3HexagonRes7', ["SHAPE@", "H3_ID"]) as cursor:
    
    for row in h3hexagons:
        
        boundary = h3.cell_to_boundary(row)
        
        array = arcpy.Array()
        
        for lat, lon in boundary:
            
            array.add(arcpy.Point(lon, lat))  # flip back
        
        polygon = arcpy.Polygon(array, spatialreference)
        
        cursor.insertRow([polygon, row])
```

```python
AlbersEqualArea = arcpy.SpatialReference(5070)  # NAD83 Conus Albers Equal Area


arcpy.management.Project('RedWingH3HexagonRes7','RedWingnResoluiton7',AlbersEqualArea)
```
